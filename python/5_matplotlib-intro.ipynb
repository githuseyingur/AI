```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# jupyter notebook'a özel tanımlamadır.
```

```python
ageList = [8,9,11,13,14,17,18,19,24,27,29,33,37,41,43,47,49,50,54,56,64,72,80,83]
intelligenceList = [84,92,100,104,112,115,120,124,132,134,138,139,144,140,130,126,122,118,116,114,112,110,104,100]

npAgeList = np.array(ageList)
npIntelligenceList = np.array(intelligenceList)

plt. plot(npAgeList,npIntelligenceList,"b")  # r : red , g : green , b : blue, y : yellow
plt.xlabel('age')
plt.ylabel('intelligence')
plt.title('Zekanın Yaşa Göre Değişim Grafiği')
plt.show()
```

```python
numpyArray1 = np.linspace(0,10,18)
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 3
numpyArray2
```

```python
plt.plot(numpyArray1,numpyArray2, 'g--')
```

```python
plt.plot(numpyArray1,numpyArray2, 'b*-')
```

```python
plt.subplot(1,2,1) # 1 satır, 2 sütun, 1.grafik
plt.plot(numpyArray1,numpyArray2,'r*-')
plt.subplot(1,2,2) # 1 satır, 2 sütun, 2.grafik
plt.plot(numpyArray2,numpyArray1,'g--')


```

### FIGURE

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.3,0.3,0.3,0.3]) # [X,Y,width,height]
figureAxes.plot(numpyArray1,numpyArray2, 'g')
figureAxes.set_xlabel('X Ekseni')
figureAxes.set_ylabel('Y Ekseni')
figureAxes.set_title('Grafik Başlığı')
```

### NESTED GRAPHS

```python
figure2 = plt.figure()

axes1 = figure2.add_axes([0.1,0.1,0.6,0.6]) # [X,Y,width,height]
axes2 = figure2.add_axes([0.2,0.4,0.2,0.2])

axes1.plot(numpyArray1,numpyArray2,'g')
axes1.set_xlabel('X Ekseni')
axes1.set_ylabel('Y Ekseni')
axes1.set_title('Ana Grafik')

axes2.plot(numpyArray2,numpyArray1,'b')
axes2.set_xlabel('X Ekseni')
axes2.set_ylabel('Y Ekseni')
axes2.set_title('Küçük Grafik')
```

### SUBPLOTS

```python
(myFigure,myAxes) = plt.subplots(nrows=1,ncols=2)

for axes in myAxes:
    axes.plot(numpyArray1,numpyArray2,'g')
    axes.set_xlabel('X Ekseni')

plt.tight_layout()  # grafikler arası mesafe koy.
    
# myFigure
```

### GÖRSEL GELİŞTİRMELER

```python
newFigure = plt.figure(figsize=(6,4), dpi = 120) # figsize = (width, height) --> axisi override eder. | dpi : kalite

newAxes = newFigure.add_axes([0.1,0.1,0.5,0.5])
newAxes.plot(numpyArray1,numpyArray1 ** 2, label = 'numpyArray ** 2')   # grafik veri etiketi
newAxes.plot(numpyArray1,numpyArray1 ** 3, label = 'numpyArray ** 3')   # grafik veri etiketi

newAxes.legend(loc=0) # labelin konumu 0 : sol köşe
```

```python
# Grafiği png KAYDET
newFigure.savefig("my_figure.png", dpi = 200)
```

### GÖRSELLİK İLERİ SEVİYE

```python
(Figure1,Axes1) = plt.subplots()
Axes1.plot(numpyArray1,numpyArray2,color='#3A95A8',alpha=0.6)
Axes1.plot(numpyArray2,numpyArray1,color='#C96F23',alpha=0.8)


```

```python
(Figure2,Axes2) = plt.subplots()
Axes2.plot(numpyArray1,numpyArray1 + 2,color='blue',linewidth = 7.0)
Axes2.plot(numpyArray1,numpyArray1 + 3,color='yellow')

Axes2.plot(numpyArray1,numpyArray1 + 4,color='pink',linestyle = '--')
Axes2.plot(numpyArray1,numpyArray1 + 5,color='brown',linestyle = '-.')
Axes2.plot(numpyArray1,numpyArray1 + 6,color='red',linestyle = ':')
Axes2.plot(numpyArray1,numpyArray1 + 6,color='black',linestyle = '-', marker = 'o', markersize = 6, markerfacecolor = 'red')
```
